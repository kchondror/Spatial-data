```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as mlp
import seaborn as sns
import geopandas as gpd
import sys
import re
import fileinput

```

```python
df = pd.read_xml("map.xml")
df.head(10)
```

```python
df = df.drop(["minlat","minlon","maxlat","maxlon","visible","version","changeset","timestamp","user","uid","tag","nd","member"], axis=1)

df = df.iloc[1:,:]
```

```python
print(df.isnull().sum())
df = df.dropna()
```

```python
df = df.applymap(lambda x: str(x))

df.head()
```

```python
def Empty_datafile():
    file_to_delete = open("datafile.txt",'w')
    file_to_delete.close()
    
Empty_datafile()
```

```python
BlockSize = 300#32768
num_records = 0 
num_blocks  = 0 
block_id = 0

Block = np.array([]) 
#print(sys.getsizeof(Block))
for i in df.iterrows():
    #print(f"size of element : {sys.getsizeof(i[1])}")
    
    
    temp = sys.getsizeof(np.append(Block,i[1]))
    if(temp <= BlockSize):
        Block = np.append(Block,i[1])
        num_records+=1
        #print(f"{Block}\n\n Size of current Block:{sys.getsizeof(Block)}")
        #input()
    else:
        #print("New Block")
        num_blocks += 1
        block_id += 1
        with open("datafile.txt", "a") as file:
            string = "Block" + str(block_id) + " " + str(Block)
            #print(f"Stirng size : {sys.getsizeof(string)}")
            file.write(string)
            file.write(", ")
            
            
            file.close()
        
        Block = np.asarray([i[1]])

def line_pre_adder(filename, line_to_prepend):
    f = fileinput.input(filename, inplace=1)
    for xline in f:
        if f.isfirstline():
            print (line_to_prepend + xline)
        else:
            print (xline)
line_pre_adder("datafile.txt" , f"Block0 [records: {num_records} blocks: {num_blocks}] ") 
```
