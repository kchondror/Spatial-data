```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as mlp
import seaborn as sns
import geopandas as gpd
import sys

```

```python
df = pd.read_xml("map.xml")
df.head(10)
```

```python
df = df.drop(["minlat","minlon","maxlat","maxlon","visible","version","changeset","timestamp","user","uid","tag","nd","member"], axis=1)

df = df.iloc[1:,:]
```

```python
print(df.isnull().sum())
df = df.dropna()
```

```python
df = df.applymap(lambda x: str(x))

df.head()
```

```python
BlockSize = 300#32768

Block = np.array([]) 
print(sys.getsizeof(Block))
for i in df.iterrows():
    print(f"size of element : {sys.getsizeof(i[1])}")
    
    
    temp = sys.getsizeof(np.append(Block,i[1]))
    if(temp <= BlockSize):
        Block = np.append(Block,i[1])
        
        print(f"{Block}\n\n Size of current Block:{sys.getsizeof(Block)}")
        input()
    else:
        print("New Block")
        
        with open("datafile.txt", "a") as file:
            string = str(Block)
            print(f"Stirng size : {sys.getsizeof(string)}")
            file.write(string)
            file.write(", ")
            file.close()
            
    
        Block = np.asarray([i[1]])
```
